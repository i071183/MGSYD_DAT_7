## Titanic

In this lab we will explore the passengers on the Titanic again.

```python
# You will want to use the pandas library, and sklearn.tree
import sklearn.tree
import pandas
import numpy
```

```python
# Read in the file titanic.csv (which is in the data/ folder)
titanic = pandas.read_csv('/Users/nerenevaz/Desktop/SYDDAT7/SYD_DAT_7/data/titanic.csv',index_col='PassengerId')
titanic
```

```python
# Since we need numeric data, add another numeric column based
# upon the "Sex" column (e.g. male=0, female=1)
titanic['SexNum'] = numpy.where(titanic['Sex'] == 'male', 1,0)
titanic
```

```python
# Either drop rows where the age isn't known, or fill with
# the mean value
titanic['Age'] = titanic['Age'].fillna((titanic['Age'].mean()))
titanic
```

```python
# Create a DecisionTreeClassifier which splits based on 
# entropy with a depth of 1
dtc = sklearn.tree.DecisionTreeClassifier(max_depth=1, criterion='entropy')
```

```python
# Try to predicted the 'Survived' column from 'Pclass', 'Age' and gender
X = titanic[['Pclass','Age','SexNum']]
y = titanic.Survived
dtc.fit(X,y)
```

```python
# Use the feature_importances_ to identify what was the most important
# criteria in survival on the Titanic
dtc.feature_importances_
```

```python
# Create a deeper DecisionTreeClassifier (e.g. max_depth=3)
dtc = sklearn.tree.DecisionTreeClassifier(max_depth=3, criterion='entropy')
dtc.fit(X,y)
```

```python
dtc.feature_importances_

```

```python
# We would like to display this as a tree. 
# import StringIO and create a StringIO object which we
# can use as a file
try:
 from StringIO import StringIO
except ImportError:
 from io import StringIO
import sklearn.tree
import IPython.display
import pydot
File_obj = StringIO()
sklearn.tree.export_graphviz(dtc, out_file=File_obj, feature_names=['Pclass','Age','SexNum'])
Graph = pydot.graph_from_dot_data(File_obj.getvalue())
IPython.display.Image(Graph.create_png())
```

```python
import sklearn.grid_search
brute_force = sklearn.grid_search.GridSearchCV(
    sklearn.tree.DecisionTreeClassifier(),
    param_grid = {
        'max_depth': [1,2,3,4,5],
        'criterion': ['gini', 'entropy']
    },
    cv=10
    )
brute_force.fit(X, y)
```

```python
brute_force.best_params_
```

```python
File_obj = StringIO()
sklearn.tree.export_graphviz(brute_force.best_estimator_, 
                             out_file=File_obj, feature_names=['Pclass','Age','SexNum'])
Graph = pydot.graph_from_dot_data(File_obj.getvalue())
IPython.display.Image(Graph.create_png())
```
